```python
import matplotlib.pyplot as plt
import numpy as np

# Define the parameter space for Grid Search (example parameters)
learning_rate = np.linspace(0.01, 0.1, 4)  # Four evenly spaced learning rates
batch_size = np.linspace(100, 400, 4)      # Four evenly spaced batch sizes

# Generate the grid of points for Grid Search
grid_search_points = np.array(np.meshgrid(learning_rate, batch_size)).T.reshape(-1, 2)

# Generate random points for Random Search (same number of points as Grid Search for comparison)
np.random.seed(42)
random_search_points = np.column_stack((np.random.uniform(0.01, 0.1, 16), np.random.uniform(100, 400, 16)))

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Grid Search plot
ax[0].scatter(grid_search_points[:, 0], grid_search_points[:, 1], color='tab:blue', label='Grid Search Points')
ax[0].set_title('Grid Search', fontsize=20)
ax[0].set_xlabel('Learning Rate', fontsize=15)
ax[0].set_ylabel('Batch Size', fontsize=15)
ax[0].grid(True)

# Random Search plot
ax[1].scatter(random_search_points[:, 0], random_search_points[:, 1], color='tab:orange', label='Random Search Points')
ax[1].set_title('Random Search', fontsize=20)
ax[1].set_xlabel('Learning Rate', fontsize=15)
ax[1].set_ylabel('Batch Size', fontsize=15)
ax[1].grid(True)

# Show the plots
plt.tight_layout()
plt.savefig("../media/hyperparam_search.jpeg", bbox_inches="tight")
plt.show()
```

```python
from PIL import Image
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Load the image
image_path = "/content/media/cat.jpeg" # Replace with the path to your image file
image = Image.open(image_path)

fig, axs = plt.subplots(figsize=(10, 5))  # Adjust the figsize to make the images smaller
axs.imshow(image)
axs.axis('off')
plt.show()
```

```python

# Convert the image to RGB if it's not already in that mode
image = image.convert('RGB')

# Convert the image to a NumPy array
image_np = np.array(image)

# Extract the R, G, B channels
r_channel = image_np[:, :, 0]
g_channel = image_np[:, :, 1]
b_channel = image_np[:, :, 2]

# Display the shapes of the channels to verify
print('R channel shape:', r_channel.shape)
print('G channel shape:', g_channel.shape)
print('B channel shape:', b_channel.shape)

# Optionally, you can save the channels as separate images
r_image = Image.fromarray(r_channel)
g_image = Image.fromarray(g_channel)
b_image = Image.fromarray(b_channel)

# r_image.save(os.path.join(REPO_DIR, 'media/cat_r_channel.jpg'))
# g_image.save(os.path.join(REPO_DIR, 'media/cat_g_channel.jpg'))
# b_image.save(os.path.join(REPO_DIR, 'media/cat_b_channel.jpg'))


# Display the images
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(r_image, cmap='gray')
axs[0].set_title('Red Channel')
axs[0].axis('off')

axs[1].imshow(g_image, cmap='gray')
axs[1].set_title('Green Channel')
axs[1].axis('off')

axs[2].imshow(b_image, cmap='gray')
axs[2].set_title('Blue Channel')
axs[2].axis('off')

plt.savefig(os.path.join(REPO_DIR, 'media/cat_channels_grid.jpg'), bbox_inches='tight')
plt.show()
```

```python

# Convert the image to RGB if it's not already in that mode
image = image.convert('RGB')

# Convert the image to a NumPy array
image_np = np.array(image)

# Extract the R, G, B channels
r_channel = image_np[:, :, 0]
g_channel = image_np[:, :, 1]
b_channel = image_np[:, :, 2]

# Create empty arrays for the colored channels
r_image_np = np.zeros_like(image_np)
g_image_np = np.zeros_like(image_np)
b_image_np = np.zeros_like(image_np)

# Assign the respective channel values
r_image_np[:, :, 0] = r_channel
g_image_np[:, :, 1] = g_channel
b_image_np[:, :, 2] = b_channel

# Convert NumPy arrays back to images
r_image = Image.fromarray(r_image_np)
g_image = Image.fromarray(g_image_np)
b_image = Image.fromarray(b_image_np)


# # Optionally, save the images
# r_image.save(os.path.join(REPO_DIR, 'media/cat_r_channel_colored.jpg'))
# g_image.save(os.path.join(REPO_DIR, 'media/cat_g_channel_colored.jpg'))
# b_image.save(os.path.join(REPO_DIR, 'media/cat_b_channel_colored.jpg'))

# Display the images
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(r_image_np)
axs[0].set_title('Red Channel')
axs[0].axis('off')

axs[1].imshow(g_image_np)
axs[1].set_title('Green Channel')
axs[1].axis('off')

axs[2].imshow(b_image_np)
axs[2].set_title('Blue Channel')
axs[2].axis('off')

plt.savefig(os.path.join(REPO_DIR, 'media/cat_channels_grid_colored.jpg'), bbox_inches='tight')
plt.show()
```
